## One Hot Encoding

### 🔎 Why One-Hot Encoding before PCA?
- PCA requires **numerical input**, but all features in the Mushroom dataset are categorical.  
- One-hot encoding converts categories into **binary (0/1) features**, avoiding false numeric order.  
- This ensures PCA captures **real variance** between categories rather than artificial distances.  
- It also keeps the dataset **unbiased and interpretable** for dimensionality reduction.  


```python
#separate target and features
y=df["class"]
X=df.drop("class",axis=1)

#One-hot encode categorical features
encoder = OneHotEncoder(drop="first",sparse_output=False)
x_encoded = encoder.fit_transform(X)

#convert to DataFrame for better readability
x_encoded = pd.DataFrame(x_encoded,columns=encoder.get_feature_names_out(X.columns))

x_encoded.head()
```

### Multicollinearity Issue 

- **`OneHotEncoder(drop=None)` (default)**  
  - Keeps **all categories** for each feature.  
  - Example: `color = {Red, Blue, Green}` → creates 3 columns: `Red`, `Blue`, `Green`.  
  - No information is lost, but introduces **multicollinearity** (one column can be derived from others).  
  - Good when using PCA or tree-based models (they are not affected much by collinearity).  

- **`drop="first"`**  
  - Drops the **first category** for each feature to avoid the **dummy variable trap**.  
  - Example: `color = {Red, Blue, Green}` → creates only 2 columns: `Blue`, `Green` (Red is implied).  
  - Reduces feature count, avoids redundancy in linear models (like Logistic Regression without regularization).  
  - But information is slightly encoded implicitly.  

✅ **In short:**  
- `drop=None` → keeps all dummy variables (more features, possible redundancy).  
- `drop="first"` → removes one dummy per feature (fewer features, avoids multicollinearity).

### Part A3 : Standardization


### 🔎 Why Standardization before PCA?
- PCA is **variance-based**: it gives more weight to features with larger variance.  
- Even though one-hot encoded features are binary (0/1), their variances differ depending on category frequency.  
- Standardization ensures that **all features contribute equally**, by scaling them to mean = 0 and standard deviation = 1.  
- Without scaling, features with high frequency (like common categories) could dominate the first principal components.  


```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler = StandardScaler()

# Fit on one-hot encoded features and transform
X_scaled = scaler.fit_transform(x_encoded)

# Convert back to DataFrame for readability
X_scaled = pd.DataFrame(X_scaled, columns=x_encoded.columns)

# Check mean and std (should be ~0 and ~1)
print("Means of first 10 features:\n", X_scaled.iloc[:, :10].mean().round(4))
print("\nStandard deviations of first 10 features:\n", X_scaled.iloc[:, :10].std().round(4))

# Show first 5 rows
X_scaled.head()
```

## Part B PCA Implementation

## Part B1 — Apply PCA
    We apply PCA to the standardized one-hot encoded dataset (X_scaled).
    We don’t specify the number of components yet, so PCA will compute them all.


```python
# Initialize PCA with all components
pca = PCA()

# Fit PCA on standardized dataset
x_pca = pca.fit_transform(x_scaled)
```

```python
# Explained variance ratio
explained_variance = pca.explained_variance_ratio_

# Cumulative variance
cumulative_variance = np.cumsum(explained_variance)
```

## Part B2 — Scree Plot & Optimal Components
The explained variance ratio tells us how much variance each principal component explains.
We plot:
- Blue bars: explained variance per PC
- Red line: cumulative explained variance
We also draw a horizontal line at 95% to decide how many PCs are needed to preserve most of the variance.


```python
# Scree plot
plt.figure(figsize=(10,5))
plt.bar(range(1, len(explained_variance)+1), explained_variance, alpha=0.6, label="Explained variance per PC")
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker="o", color="red", label="Cumulative explained variance")
plt.axhline(y=0.95, color="gray", linestyle="--", linewidth=1, label="95% threshold")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot & Cumulative Variance")
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```

```python
# Number of components for 95% variance
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Number of components to retain 95% variance: {n_components_95}")
```

### 📌 Number of Components to Retain 95% Variance

- PCA reduces dimensionality while preserving as much variance (information) as possible.  
- From the scree plot and cumulative variance analysis, we find that **55 principal components** are required to explain **95% of the total variance** in the mushroom dataset.  
- This means instead of using all one-hot encoded features (~100+), we can compress the data into **55 components** without losing more than 5% of the information.  
- Although 55 is still large, it is **much smaller than the original feature space**, reducing redundancy and multicollinearity while keeping most of the useful structure.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import ipywidgets as widgets
from ipywidgets import interact

# Define colorblind-friendly colormap (blue vs orange)
from matplotlib.colors import ListedColormap
cmap_custom = ListedColormap(["#0072B2", "#E69F00"])  # blue, orange

def plot_pca_decision_boundary(pc_x=1, pc_y=2):
    # Select PCs
    X_pca2 = x_pca[:, [pc_x-1, pc_y-1]]
    lr = LogisticRegression(max_iter=2000)
    lr.fit(X_pca2, y)
    
    y_pred = lr.predict(X_pca2)
    acc = accuracy_score(y, y_pred)

    # Grid
    x_min, x_max = X_pca2[:, 0].min() - 1, X_pca2[:, 0].max() + 1
    y_min, y_max = X_pca2[:, 1].min() - 1, X_pca2[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    # Predict on grid
    Z = lr.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = np.where(Z == 'e', 0, 1).reshape(xx.shape)

    # Plot
    plt.figure(figsize=(8,6))
    plt.contourf(xx, yy, Z, alpha=0.25, cmap=cmap_custom)

    plt.scatter(
        X_pca2[:,0], X_pca2[:,1], 
        c=y.map({"e":0,"p":1}), 
        cmap=cmap_custom, 
        s=40, alpha=0.8, edgecolors="k", linewidth=0.3
    )

    # Legend
    handles = [
        plt.Line2D([], [], marker='o', color='w', markerfacecolor="#0072B2", markersize=8, label="Edible"),
        plt.Line2D([], [], marker='o', color='w', markerfacecolor="#E69F00", markersize=8, label="Poisonous")
    ]
    plt.legend(handles=handles, title="Class")

    plt.xlabel(f"PC{pc_x}")
    plt.ylabel(f"PC{pc_y}")
    plt.title(f"Decision Boundary (PC{pc_x} vs PC{pc_y})\nAccuracy = {acc:.4f}")
    plt.show()

# Interactive widget: choose any two PCs from 1 to 10
interact(plot_pca_decision_boundary,
         pc_x=widgets.IntSlider(min=1, max=10, step=1, value=1, description="PC X"),
         pc_y=widgets.IntSlider(min=1, max=10, step=1, value=2, description="PC Y"));
```

When PCs are added **cumulatively**, accuracy increases further since more variance and finer details are preserved, leading to better separation of edible and poisonous mushrooms.

- Later PCs (PC6–PC10) explain much less variance and also show high overlap, so they add little value for classification.
- But we will get a much clearer idea with the next plot.


```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KernelDensity
import matplotlib.pyplot as plt

def compute_kde_overlap(X_pca, y, n_components=10, bandwidth=0.3, n_points=1000):
    """
    Compute KDE overlap between edible and poisonous classes 
    for the first n_components principal components.
    
    Returns:
        dict: {PC name: overlap value}
    """
    results = {}
    X = pd.DataFrame(X_pca[:, :n_components], columns=[f"PC{i+1}" for i in range(n_components)])
    X["class"] = np.array(y)  # ensure numpy array
    
    for i in range(n_components):
        pc = f"PC{i+1}"
        edible = X.loc[X["class"]=="e", pc].values[:, None]
        poisonous = X.loc[X["class"]=="p", pc].values[:, None]

        if len(edible) == 0 or len(poisonous) == 0:
            results[pc] = np.nan
            continue

        # Fit KDEs
        kde_e = KernelDensity(bandwidth=bandwidth).fit(edible)
        kde_p = KernelDensity(bandwidth=bandwidth).fit(poisonous)

        # Grid for evaluation
        x_min = min(edible.min(), poisonous.min()) - 1
        x_max = max(edible.max(), poisonous.max()) + 1
        grid = np.linspace(x_min, x_max, n_points)[:, None]

        # Densities
        dens_e = np.exp(kde_e.score_samples(grid))
        dens_p = np.exp(kde_p.score_samples(grid))

        # Normalize PDFs
        dens_e /= np.trapezoid(dens_e, grid[:,0])
        dens_p /= np.trapezoid(dens_p, grid[:,0])

        # Overlap coefficient
        overlap = np.trapezoid(np.minimum(dens_e, dens_p), grid[:,0])
        results[pc] = float(overlap)  # convert np.float64 → float
    
    return results

# ---- Run for first 10 PCs ----
overlaps = compute_kde_overlap(x_pca, y, n_components=10)

# Print clean results (rounded to 4 decimals)
overlaps_clean = {k: round(v, 4) for k, v in overlaps.items()}
print(overlaps_clean)

# ---- Barplot of overlaps ----
plt.figure(figsize=(10,5))
plt.bar(overlaps_clean.keys(), overlaps_clean.values(), color="skyblue", edgecolor="k")
plt.axhline(0.5, color="red", linestyle="--", label="50% overlap")
plt.ylabel("KDE Overlap (0 = good separation, 1 = poor separation)")
plt.title("Class Overlap per Principal Component")
plt.legend()
plt.show()
```

### KDE Overlap Analysis of Principal Components  

We computed the overlap between edible and poisonous classes using KDE for the first 10 PCs.  
The values range from **0 (perfect separation)** to **1 (complete overlap)**:  

- **PC1 → 0.21**: very low overlap, strongest class separation.  
- **PC3–PC5 → ~0.33**: moderate separation, still useful for classification.  
- **PC2 (0.58), PC7 (0.56), PC8 (0.48)**: higher overlap, weaker separation.  
- **PC6 (0.69), PC9 (0.81), PC10 (0.65)**: poor separation, carry little discriminative power.  

**Conclusion:**  
The leading PCs (especially **PC1**) capture most of the discriminative variance between edible and poisonous mushrooms. Higher PCs contain more noise and redundancy, contributing less to class separation.  


# Part C  Performance Evaluation with Logistic Regression

## Part C1 — Baseline Model (Original Features)
We split the one-hot encoded, standardized dataset into training and testing sets.
Then, we train a Logistic Regression classifier and evaluate its performance
on the test set using classification metrics (precision, recall, F1-score, accuracy).


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Train-test split (stratified to keep class balance)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# Train logistic regression (baseline)
lr_baseline = LogisticRegression(max_iter=2000, random_state=42)
lr_baseline.fit(X_train, y_train)

# Predictions
y_pred_base = lr_baseline.predict(X_test)

# Evaluation
print("Baseline Model (Original standardized features)")
print(classification_report(y_test, y_pred_base))
print("Accuracy:", accuracy_score(y_test, y_pred_base))
```

### Baseline Model: Logistic Regression on Original Features  

We trained a Logistic Regression classifier on the **original standardized features** (without PCA).  
The model achieved a **perfect classification**:  

- **Accuracy:** 100%  
- **Precision, Recall, F1-score:** 1.00 for both edible and poisonous classes  

This indicates that the original high-dimensional feature space already provides complete separability between edible and poisonous mushrooms.  
The baseline thus serves as the upper bound for comparison when evaluating PCA-transformed models.  


## Part C2 — PCA-Transformed Model
We now transform both train and test sets using the optimal number of components
determined in Part B (e.g., enough to capture 95% variance).
We then train and evaluate Logistic Regression on these reduced-dimension features.


```python
from sklearn.decomposition import PCA

# Use optimal number of components from Part B
pca_final = PCA(n_components=n_components_95, random_state=42)

# Fit PCA on training data only (to avoid leakage)
X_train_pca = pca_final.fit_transform(X_train)
X_test_pca = pca_final.transform(X_test)

# Train logistic regression on PCA data
lr_pca = LogisticRegression(max_iter=2000, random_state=42)
lr_pca.fit(X_train_pca, y_train)

# Predictions
y_pred_pca = lr_pca.predict(X_test_pca)

# Evaluation
print("PCA Model (Reduced dimensions)")
print(classification_report(y_test, y_pred_pca))
print("Accuracy:", accuracy_score(y_test, y_pred_pca))
```

### PCA-Transformed Model: Logistic Regression on Reduced Dimensions  

After applying PCA and retaining the optimal number of components (capturing ~95% of the variance),  
we trained Logistic Regression on the reduced feature space.  

- **Accuracy:** ~99.88%  
- **Precision, Recall, F1-score:** all very close to 1.00 for both classes  

This shows that even after dimensionality reduction, the model maintains almost the same performance as the baseline.  
PCA successfully compressed the high-dimensional dataset into fewer components while preserving nearly all of the discriminative information.  


```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import ipywidgets as widgets
from ipywidgets import interact
import numpy as np

# Precompute accuracies for all components
max_components = 50  # you can increase up to X_scaled.shape[1]
acc_scores = []

for k in range(1, max_components+1):
    pca = PCA(n_components=k, random_state=42)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    lr = LogisticRegression(max_iter=2000, random_state=42)
    lr.fit(X_train_pca, y_train)
    y_pred = lr.predict(X_test_pca)
    acc = accuracy_score(y_test, y_pred)
    acc_scores.append(acc)

acc_scores = np.array(acc_scores)

# Interactive function
def plot_accuracy(n_components=10):
    plt.figure(figsize=(8,6))
    plt.plot(range(1, max_components+1), acc_scores, marker='o', label="Accuracy")
    plt.axvline(x=n_components, color='r', linestyle='--', label=f"Selected: {n_components}")
    plt.axhline(y=acc_scores[n_components-1], color='g', linestyle='--')
    plt.scatter(n_components, acc_scores[n_components-1], color="red", s=100, zorder=5)

    plt.xlabel("Number of PCA Components")
    plt.ylabel("Accuracy")
    plt.title("Accuracy vs Number of PCA Components")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Accuracy with {n_components} components: {acc_scores[n_components-1]:.4f}")

# Slider to select number of components
interact(plot_accuracy, n_components=widgets.IntSlider(min=1, max=max_components, step=1, value=10));
```

```python
# Assuming your standardized features are stored in X_scaled
pca = PCA()
pca.fit(X_scaled)

# Explained variance (cumulative)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

def variance_retained(n_components):
    if n_components < 1 or n_components > len(cumulative_variance):
        raise ValueError(f"n_components must be between 1 and {len(cumulative_variance)}")
    return cumulative_variance[n_components-1]

# Example usage
print("Variance retained with 21 components:", variance_retained(21))
```

### Accuracy vs Number of PCA Components  

The plot shows how Logistic Regression accuracy changes as the number of principal components increases:  

- With very few components (< 10), accuracy is relatively low (~87–96%).  
- Accuracy improves sharply as more components are added, reaching ~99.9% around **21 components**.  
- Beyond 21 components, accuracy plateaus, indicating no further performance gain. Although only ~67% variance is retained (vs ~95% with 55 PCs), these components capture the class-separating information while removing redundancy.
- 
**Conclusion:**  
Instead of using all 55 features after one-hot encoding, we can retain only **21 principal components** to achieve the same accuracy (~99.88%) as the full dataset.  
This demonstrates the effectiveness of PCA in reducing dimensionality while preserving classification performance.  


## Part C3 — Comparison and Analysis

```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score

# Collect metrics into a table
results = []
for name, pred in [("Baseline", y_pred_base), ("PCA", y_pred_pca)]:
    acc = accuracy_score(y_test, pred)
    prec = precision_score(y_test, pred, average="weighted", zero_division=0)
    rec = recall_score(y_test, pred, average="weighted", zero_division=0)
    f1 = f1_score(y_test, pred, average="weighted", zero_division=0)
    results.append([name, acc, prec, rec, f1])

comparison_df = pd.DataFrame(results, columns=["Model", "Accuracy", "Precision", "Recall", "F1-score"])
print("Comparison of Logistic Regression models:")
display(comparison_df)
```
